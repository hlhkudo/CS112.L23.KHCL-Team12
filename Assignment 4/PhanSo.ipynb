# Phân số (1s, 50MB)

Cho 2 phân số đúng và tối giản $\frac{a}{b}$, $\frac{c}{d}$. Mỗi phép biến đổi là tăng ***a*** và ***b*** lên 1, sau đó giản ước phân số nhận được. 

Hãy xác định sau bao nhiêu bước biến đổi từ phân số thứ nhất ban đầu nhận được phân số thứ hai đã cho.

***Dữ liệu:*** Vào từ thiết bị nhập chuẩn gồm 4 dòng, mỗi dòng chứa một số nguyên ***a***, ***b***, ***c***, ***d***, $0 < a < b ≤ 10^5$, $0 < c < d ≤ 10^5$, ***a*** và ***b*** nguyên tố cùng nhau, ***c*** và ***d*** nguyên tố cùng nhau $\frac{a}{b}$ ≠ $\frac{c}{d}$.

***Kết quả:*** Đưa ra thiết bị xuất chuẩn số 0 nếu không có cách biến đổi hoặc một số nguyên – số lượng phép biến đổi.

***Ví dụ:***

\begin{array}{|c|c|}
INPUT&OUTPUT\\ \hline
1 \ 6 \ 2 \ 3 & 5
\end{array}

## Abstraction

* Đếm số lần biến đổi của phân số thứ nhất sao cho bằng phân số thứ hai với mỗi lần biến đổi là cộng thêm tử và mẫu cho 1 sau đó rút gọn phân số nếu có thể.

## Pattern recognition

* Brute force 

## Algorithm design

* ***Ý tưởng:***

  - Dùng brute force để tính từng trường hợp thỏa điều kiện đề bài, nếu đúng thì cộng biến đếm thêm 1, cuối cùng in ra biến đếm. Vậy lúc này, câu hỏi đặt ra là chúng ta sẽ chạy vòng lặp với điều kiện là gì?
  - Dễ dàng nhận thấy nếu $\frac{a}{b}$ < $\frac{c}{d}$ thì vẫn còn có thể biến đổi để $\frac{a}{b}$ = $\frac{c}{d}$, nhưng một khi $\frac{a}{b}$ > $\frac{c}{d}$ rồi thì cho dù có biến đổi $\frac{a}{b}$ cách mấy cũng không thể nào bằng $\frac{c}{d}$ được nữa. Vì vậy mà điều kiện để chạy vòng lặp sẽ là $\frac{a}{b}$ < $\frac{c}{d}$
  
* ***Giải thích thuật toán:***

  - Khởi tạo biến đếm count = 0.

  - Chạy vòng lặp với điều kiện là $\frac{a}{b}$ < $\frac{c}{d}$:

    - Cộng thêm 1 cho a và b 

    - Cộng thêm 1 cho biến count

    - Tìm ước chung lớn nhất (UCLN) của a và b

    - Chia a và b cho UCLN vừa tìm được và lấy phần nguyên của kết quả phép chia để rút gọn phân số
    - Đến khi nào điều kiện sai thì thoát khỏi vòng lặp
  
  - Nếu $\frac{a}{b}$ > $\frac{c}{d}$ thì xuất ra màn hình số 0. Ngược lại, xuất ra màn hình giá trị biến count

* ***Lưu đồ thuật toán:***

* ***Chú ý:***
  
  - Trong quá trình làm bài thì có 1 vấn đề đã xảy ra là code của nhóm em bị **Runtime Error** và sau nhiều lần sửa code thì nhóm em phát hiện ra vấn đề như sau:
    - Input đầu vào bao gồm 4 dòng, mỗi dòng chứa 1 số nguyên
    - Trong khi đó, cả nhóm em lại quá chăm chú vào phần ví dụ với hình minh họa lại có Input là 1 dòng gồm 4 số nguyên
  - **Kết luận:** hãy đọc kĩ đề bài và chỉ tin vào yêu cầu của đề, đừng quá tin vào hình ảnh "***chỉ mang tính chất minh họa***".

## Complexity evaluation

* Time complexity: $O(n^2)$

## Programming

```python
%%time
def ucln(a, b):
    if (b == 0):
        return a
    return ucln(b, a % b)
a = int(input())
b = int(input())
c = int(input())
d = int(input())
count = 0
while a/b < c/d:
    a += 1
    b += 1
    count += 1
    e = ucln(a, b)
    a = a // e
    b = b // e
if a/b > c/d:
    print(0)
else:
    print(count)
```
